```python
import numpy as np
import matplotlib.pyplot as plt

# Analytical solution of the differential equation f'(t) = f(t) with f(0) = 1
def analytical_solution(t):
    return np.exp(t)

# Euler's method
def euler_method(f, t0, y0, h, n):
    t_values = [t0]
    y_values = [y0]
    for i in range(n):
        y0 = y0 + h * f(t0, y0)
        t0 = t0 + h
        t_values.append(t0)
        y_values.append(y0)
    return t_values, y_values

# Differential equation f'(t) = f(t)
def f(t, y):
    return y

# Initial conditions
t0 = 0
y0 = 1
h = 0.5
n = 3

# Apply Euler's method
t_values, y_values = euler_method(f, t0, y0, h, n)

# Calculate true values
true_values = [analytical_solution(t) for t in t_values]

# Print the results
for i in range(len(t_values)):
    print(f"t = {t_values[i]:.1f}, Euler's method = {y_values[i]:.5f}, True value = {true_values[i]:.5f}")

# Plot the results
plt.plot(t_values, y_values, 'o-', label="Euler's method")
plt.plot(t_values, true_values, 's-', label="True value")
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend()
plt.show()
```
